```python
import time
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
audio_df = pd.read_csv(r"HCfeatures.csv", header = None)
video_df = pd.read_csv(r"video11tasks_aus_gaze_mean_std.csv", header = None)
truth_df = pd.read_csv(r"binary_stress_2class.csv")
```

```python
audio_df = audio_df.drop(columns=[audio_df.columns[0]])
video_df = video_df.drop(columns=[video_df.columns[0]])

```

```python
video_df = video_df.dropna()
```

```python
video_df = video_df.drop(index=0)
```

```python
video_df
```

```python
# Extract the labels from truth.csv (assuming labels are in the second column)
y = truth_df.iloc[:, 1]  # Second column of truth.csv for labels
```

```python
y
```

Balacing the data using smote upscaler

```python
# Create an instance of SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
```

```python
# Apply SMOTE to generate synthetic samples
X_resampled, y_resampled = smote.fit_resample(audio_df, y)
```

```python
X_resampled
```

```python
y_resampled
```

merging the audio and video data 

```python
merged_df = pd.concat([X_resampled, video_df], axis=1)
```

```python
merged_df
```

```python
merged_df = merged_df.dropna()
```

```python
merged_df
```

```python
y_resampled
```

```python
y_resampled = y_resampled.drop(index=0).reset_index(drop=True)
```

```python
y_resampled
```

```python
X_train, X_test, y_train, y_test = train_test_split(merged_df, y_resampled, test_size=0.2, random_state=42)
```

classification using the merged data

```python
list_classif = [
    RandomForestClassifier(max_depth=5, random_state=0), 
    KNeighborsClassifier(n_neighbors=3), 
    SVC(gamma='auto', kernel='rbf'), 
    MLPClassifier(max_iter=5000, random_state=0, hidden_layer_sizes=(100,))
]
```

```python
for classifier in list_classif:
    # Measure training time
    start_time = time.time()
    
    # Fit the classifier
    classifier.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = classifier.predict(X_test)
    
    # Measure training duration
    training_time = time.time() - start_time
    
    # Calculate accuracy and F1 score
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Print results
    print(f"Classifier: {classifier.__class__.__name__}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Training Time: {training_time:.4f} seconds")
    print(classification_report(y_test, y_pred))
```
